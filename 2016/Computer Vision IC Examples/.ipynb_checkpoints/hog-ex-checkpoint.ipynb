```python
import numpy as np
import scipy as sp
from scipy import ndimage
from scipy import signal
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
tri_bgr = cv2.imread('tri.png')
tri = cv2.cvtColor(tri_bgr, cv2.COLOR_BGR2GRAY)

cir_bgr = cv2.imread('circle.png')
cir = cv2.cvtColor(cir_bgr, cv2.COLOR_BGR2GRAY)

sq_bgr = cv2.imread('square.png')
sq = cv2.cvtColor(sq_bgr, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10,5))
plt.subplot(131)
plt.imshow(tri, cmap='gray')
plt.title('Triangle')
plt.subplot(132)
plt.imshow(sq, cmap='gray')
plt.title('Square')
plt.subplot(133)
plt.imshow(cir, cmap='gray')
plt.title('Circle')
```

```python
dx = np.array([[-1, 1]])

print dx.shape
print tri.shape

tri_dx = sp.signal.convolve(tri, dx, mode='same')
tri_dy = sp.signal.convolve(tri, dx.T, mode='same')
tri_mag = np.sqrt(np.power(tri_dx, 2) + np.power(tri_dy, 2))
tri_ori = np.arctan2(tri_dy, tri_dx)

plt.figure(figsize=(20,20))
plt.subplot(411)
plt.imshow(tri_dx, cmap='gray')
plt.subplot(412)
plt.imshow(tri_dy, cmap='gray')
plt.subplot(413)
plt.imshow(tri_mag, cmap='gray')
plt.subplot(414)
plt.imshow(tri_ori)
```
